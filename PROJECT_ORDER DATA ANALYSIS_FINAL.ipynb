```python
df.dropna()
df.dropna(inplace=True)
df
```

```python
df.rename(columns = {'Order Id': 'Order_Id', 'Order Date': 'Order_Date' , 'Ship Mode': 'Ship_Mode', 'Postal Code': 'Postal_Code', 'Sub Category': 'Sub_Category',
       'Product Id': 'Product_Id', 'cost price': 'Cost_Price', 'List Price': 'List_Price',
       'Discount Percent': 'Discount_Percent'}, inplace = True)
df
```

```python
df['Discount'] = df['List_Price'] * (df['Discount_Percent'] / 100).round(2)

df['Sales'] = (df['List_Price'] - df['Discount']) * df['Quantity'].round(2)

df['Profit'] = df['Sales'] - (df['Cost_Price'] * df['Quantity']).round(2)
```
